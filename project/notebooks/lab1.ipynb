```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

# Загрузка MNIST из TensorFlow

```python
mnist = tf.keras.datasets.mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

# Объединяем обучающую и тестовую выборки для анализа

```python
X = np.concatenate([X_train, X_test])
y = np.concatenate([y_train, y_test])
```

# Подсчёт количества примеров для каждого класса

```python
class_counts = np.bincount(y, minlength=10)
```

# Вывод распределения классов

```python
plt.figure(figsize=(8, 6))
plt.bar(range(10), class_counts, color='skyblue', edgecolor='black')
plt.title('Распределение классов в датасете MNIST', fontsize=16)
plt.xlabel('Класс', fontsize=14)
plt.ylabel('Количество примеров', fontsize=14)
plt.xticks(range(10), fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Проверка сбалансированности

```python
total_samples = len(y)
print("Общее количество примеров:", total_samples)
for digit, count in enumerate(class_counts):
    percentage = (count / total_samples) * 100
    print(f"Класс {digit}: {count} примеров ({percentage:.2f}%)")
```
